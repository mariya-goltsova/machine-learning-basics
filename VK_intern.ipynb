```python
import lightgbm

dtrain = lightgbm.Dataset(X_train, label=y_train)
dtest = lightgbm.Dataset(X_test, label=y_test)

lgbm = lightgbm.train({'num_leaves':70, 'max_depth':8}, dtrain)
```

```python
labels_predicted_lgbm = lgbm.predict(X_test)
```

```python
labels_predicted_lgbm  = np.round(labels_predicted_lgbm )
```

```python

def dcg_score(y_true, y_score, k=10, gains="exponential"):
    order = np.argsort(y_score)[::-1]
    y_true = np.take(y_true, order[:k])

    if gains == "exponential":
        gains = 2 ** y_true - 1
    elif gains == "linear":
        gains = y_true
    else:
        raise ValueError("Invalid gains option.")

    # highest rank is 1 so +2 instead of +1
    discounts = np.log2(np.arange(len(y_true)) + 2)
    return np.sum(gains / discounts)

def ndcg_score(y_true, y_score, k=10, gains="exponential"):

    best = dcg_score(y_true, y_true, k, gains)
    actual = dcg_score(y_true, y_score, k, gains)
    return actual / best

#source: https://gist.github.com/mblondel/7337391
```

```python

def get_nDCG_score(queries_id, queries_lines_info, test_queries_lines_info, labels_true, labels_predicted):
    nDCG_scores = [] # nDCG по каждому запросу
    query_lines_in_testdata = []
    for query in queries_id:
        query_lines = queries_lines_info[query]

        query_lines_in_testdata = [elem in query_lines for elem in test_queries_lines_info]

        query_labels_true = y[query_lines].to_numpy()
        query_labels_pred = labels_predicted[query_lines_in_testdata]
        nDCG = ndcg_score(query_labels_true, query_labels_pred, k=10)
        nDCG_scores.append(nDCG)

    nDCG_Queries = np.sum(nDCG_scores) / len(queries_id) # усредняем по всем запросам
    return nDCG_Queries
```

```python
score_lgbm = get_nDCG_score(
    queries_id = queries_id_test,
    queries_lines_info = dict_query_rows,
    test_queries_lines_info = test_queries_lines_info,
    labels_true = y_test,
    labels_predicted = labels_predicted_lgbm
    )

score_lgbm
```
